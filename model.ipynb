```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,f1_score,classification_report
```

```python
df = pd.read_csv("C:\\Users\\ky805\\Downloads\\Manufacturing_defects\\manufacturing_defect_dataset.csv")
```

```python
df.head()
```

```python
df.isnull().value_counts()
```

```python
df.dropna(axis=1)
```

```python
X = df.drop(columns='DefectStatus',axis=1)
Y = df['DefectStatus']
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
logistic_model = LogisticRegression()
```

```python
decision_model = DecisionTreeRegressor()
```

```python
logistic_model.fit(X_train,Y_train)
```

```python
decision_model.fit(X_train,Y_train)
```

```python
logistic_preds = logistic_model.predict(X_test)
```

```python
decision_tree_preds = decision_model.predict(X_test)
```

```python
logistic_accuracy = accuracy_score(logistic_preds,Y_test)
print(logistic_accuracy)
```

```python
decision_accuracy = accuracy_score(decision_tree_preds,Y_test)
print(decision_accuracy)
```

```python
f1_logistic = f1_score(logistic_preds,Y_test)
print(f1_logistic)
```

```python
f1_decision = f1_score(decision_tree_preds,Y_test)
print(f1_decision)
```

```python
clasific_logistic = classification_report(Y_test,logistic_preds)
print(clasific_logistic)
```

```python
clasific_decision = classification_report(Y_test,decision_tree_preds)
print(clasific_decision)
```
